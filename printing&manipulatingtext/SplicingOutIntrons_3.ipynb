## ***Splicing out introns, part three***

Using the data from part one, write a program that will print out the original genomic DNA sequence with coding bases in uppercase and non-coding bases in lowercase. 

```python
dna = """ATCGATCGATCGATCGACTGACTAGTCATAGCTATGCATGTAGCTACTCGATCGATCGA
TCGATCGATCGATCGATCGATCGATCATGCTATCATCGATCGATATCGATGCATCGACT
ACTAT"""

# Remove newline characters
new_dna = dna.replace("\n", "")

# exon1: 1st to 63rd character  => slice 0:63
exon1 = new_dna[0:62]

# exon2: 91st to last character => slice 90:end
exon2 =new_dna[90:len(new_dna)]

#non-coding region (intron): 63rd to 90th character => slice 62:90
intron = new_dna[62:90].lower()

#print out the original genomic DNA sequence with coding bases in uppercase & non-coding bases in lowercase. 

print (f"The genomic DNA sequence is {exon1}{intron}{exon2}")
```
