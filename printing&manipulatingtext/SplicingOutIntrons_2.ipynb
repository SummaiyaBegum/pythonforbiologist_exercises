### ***Splicing out introns, part two***

Using the data from part one, write a program that will calculate what percentage
of the DNA sequence is coding

```python
dna = """ATCGATCGATCGATCGACTGACTAGTCATAGCTATGCATGTAGCTACTCGATCGATCGA
TCGATCGATCGATCGATCGATCGATCATGCTATCATCGATCGATATCGATGCATCGACT
ACTAT"""

# Remove newline characters
new_dna = dna.replace("\n", "")

# exon1: 1st to 63rd character  => slice 0:63
exon1 = new_dna[0:62] 

# exon2: 91st to last character => slice 90:end
exon2 =new_dna[90:len(new_dna)]

coding_region = exon1 + exon2

#calculate what percentage of the DNA sequence is coding

percentage_coding = (len(coding_region) / len(new_dna)) * 100

print(f"The percentage of coding DNA is {percentage_coding:.2f}%")
```
