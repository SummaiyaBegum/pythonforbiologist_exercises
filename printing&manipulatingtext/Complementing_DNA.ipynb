## ***Complementing DNA***

Here's a short DNA sequence:

`ACTGATCGATTACGTATAGTATTTGCTATCATACATATATATCGATGCGTTCAT`

Write a program that will print the complement of this sequence.

```python
dna = "ACTGATCGATTACGTATAGTATTTGCTATCATACATATATATCGATGCGTTCAT"

comp_dna = dna.replace("A", "t").replace("T", "a").replace("C", "g").replace("G", "c").upper()

print(f"The complementary DNA is {comp_dna}")
```

```python
#Using Biopython

from Bio.Seq import Seq

dna = Seq("ACTGATCGATTACGTATAGTATTTGCTATCATACATATATATCGATGCGTTCAT")
comp_dna = dna.complement()

print(f"The complementary DNA is {comp_dna}")
```
