###**Splicing out introns, part one**

Here's a short section of genomic DNA:

`ATCGATCGATCGATCGACTGACTAGTCATAGCTATGCATGTAGCTACTCGATCGATCGA`
`TCGATCGATCGATCGATCGATCGATCATGCTATCATCGATCGATATCGATGCATCGACT`
`ACTAT`

It comprises two exons and an intron. The first exon runs from the start of the sequence to the sixty-third character, and the second exon runs from the ninetyfirst character to the end of the sequence. Write a program that will print just the coding regions of the DNA sequence. 

```python
dna = """ATCGATCGATCGATCGACTGACTAGTCATAGCTATGCATGTAGCTACTCGATCGATCGA
TCGATCGATCGATCGATCGATCGATCATGCTATCATCGATCGATATCGATGCATCGACT
ACTAT"""

# Remove newline characters
new_dna = dna.replace("\n", "")

# exon1: 1st to 63rd character  => slice 0:63
exon1 = new_dna[0:62] 

# exon2: 91st to last character => slice 90:end
exon2 =new_dna[90:len(new_dna)]

coding_region = exon1 + exon2

print(f"The coding regions of the DNA sequence {coding_region}")

```
